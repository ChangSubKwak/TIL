```python
import matplotlib.pyplot as plt
from glob import glob
import numpy as np

imageList = []
yList = []
images = glob('./img/*')
resiziedImage = ''
for image in images[:2]:
    tempImage = Image.open(image)
    tempImage = tempImage.resize((30,40))
    imageArray = np.array(tempImage) / 255
    imageList.append(imageArray)
    yList.append(0)

print(yList)
print(np.shape(imageList))
print(np.max(imageList))
np.savez('xyData.npz', x=imageList, y=yList)

loadData = np.load('xyData.npz')
xData = loadData['x']
yData = loadData['y']
print(xData.shape)
print(yData.shape)
print(list(loadData))
# loadData
```
