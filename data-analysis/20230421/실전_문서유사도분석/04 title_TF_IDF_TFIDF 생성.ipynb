# 1. title 단어 불러오기

```python
# 질문 내용 단어 추출 pickle 가져오기
with open('./data/nouns_title.pickle', 'rb') as f:
    nouns = pickle.load(f)
```

```python
len(nouns)
```

# 2. 단어 필드 제목 _ 단어 unique

```python
wordList=[]
vocab=sorted(list(set([n for i in nouns for n in i])))
vocab[:20]
```

```python
#한글자음, 한자 삭제,특수 문자
tmp=[''.join(re.findall('[가-힣a-zA-Z0-9]+',i)) for i in vocab]
vocab=[i for i in sorted(list(set(tmp))) if i!='']
vocab[:20]
```

# 3. TF

```python
dtm=[]
for i in nouns:
    dtm.append([])
    for j in vocab:
        dtm[-1].append(i.count(j))
tf_df=pd.DataFrame(dtm, columns=vocab)
```

```python
tf_df
```

```python
#tf_df pickle로 저장
import pickle
with open('./data/titleTF.pickle', 'wb') as f:
    pickle.dump(tf_df, f, pickle.HIGHEST_PROTOCOL)
```

# 4. IDF

```python
D=len(tf_df) #총문서수
df=tf_df.astype(bool).sum()
idf=np.log10(D/(df+1))
idf
```

```python
#idf pickle로 저장
with open('./data/titleIDF.pickle', 'wb') as f:
    pickle.dump(idf, f, pickle.HIGHEST_PROTOCOL)
```

# 5. TF_IDF

```python
tfidf=tf_df*idf
tfidf
```

```python
#tfidf_ pickle로 저장
with open('./data/titleTFIDF.pickle', 'wb') as f:
    pickle.dump(tfidf, f, pickle.HIGHEST_PROTOCOL)
```
