```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm

image = Image.open("dogs.jpg")
imageArray = np.array(image)

lastIndex = 5

# print(type(image))
# print(type(imageArray))
print(np.max(imageArray), np.min(imageArray))
# print(imageArray.shape)
# print(imageArray[0].shape)
# print(imageArray[0][0])

##### 1
original = imageArray
print(np.max(original), np.min(original))

plt.subplot(1, lastIndex, 1)
plt.imshow(original)

##### 2
divide255 = imageArray / 255
print(np.max(divide255), np.min(divide255))

plt.subplot(1, lastIndex, 2)
plt.imshow(divide255)

##### 3
divide127 = imageArray / 127.5
print(np.max(divide127), np.min(divide127))

plt.subplot(1, lastIndex, 3)
plt.imshow(divide127)

##### 4
divide127AndSubTractOne = imageArray / 127.5 - 1
print(np.max(divide127AndSubTractOne), np.min(divide127AndSubTractOne))

plt.subplot(1, lastIndex, 4)
plt.imshow(divide127AndSubTractOne)

# extent
plt.subplot(1, lastIndex, 5)
plt.imshow(original, cm.viridis, extent=(0, 1, 0, 1))
# plt.imshow(original, cm.viridis, extent=(-1, 1, -1, 1))


# Pose Estimation with mediapipe

```
